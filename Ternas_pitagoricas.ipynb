Una terna pitagorica $(x,y,z)$ son aquellos numeros tales que $x^2 + y^2 = z^2$

Usando comprensiones intentaremos mostrar todas  ternas pitagoricas tales que
$x,y,z\leq 10^3$

```python
ternas = [[x,y,x**2 + y**2] for x in range(1,10**3 + 1) for y in range(1,10**3 + 1) if x**2 + y**2 <= 1000]
print(ternas)
```
